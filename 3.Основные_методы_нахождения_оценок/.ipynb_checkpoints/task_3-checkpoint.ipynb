```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline

data = []

file = open('Weibull.csv', 'r')
for str in file:
    for c in str.split():
        data.append(float(c))
file.close()
```

1) Дифференцируем функцию распределения, находим плотность: 
$ p(x) = \gamma \cdot x^{\gamma - 1} e^{-x^{\gamma}} $
    
2) Функция правдоподобия:
$ f = \gamma^N \prod_{i=1}^{N} e^{-x^{\gamma}_i} \cdot x_i^{\gamma - 1} = \gamma^N e^{- \sum^N_{i=1} x^{\gamma}_i} \prod_{i=1}^{N} x_i^{\gamma - 1} $

3) Логарифм от функции правдоподобия:
$ L = N\log \gamma - \sum_{i=1}^{N} x_i^{\gamma} + (\gamma - 1) \cdot \sum_{i=1}^{N}\log x_i $

Максимизируем логарифмическую функцию правдоподобия.

```python
x = np.arange(0.01, 100, 0.025)
```

```python
# n - количество элементов из выборки
# метод возвращает функцию правдоподобия
def estimate(n):
    L = np.zeros(len(x))

    for i in range(len(x)):
        for k in data[0:n]:
            if k == 0:
                k = 0.0001 
            L[i] += np.log(x[i]) + (x[i] - 1) * np.log(k) - (k ** x[i]) 
    return L
```

Оценим параметр формы по данным выборки за первые 4 года

```python
L = estimate(1461)
```

```python
plt.plot(x, L)
plt.ylim([-5000, 0])
plt.show()
```

```python
plt.plot(x, L)
plt.ylim([-1178, -1174])
plt.xlim([-0.34, -0.26])
plt.show()

arg_max_1 = x[np.where(L == max(L))]
print("Значение максимума функции достигается в точке ", arg_max_1)
```

Оценим параметр сдвига по полной выборке за 10 лет

```python
L = estimate(3652)
```

```python
plt.plot(x, L)
plt.ylim([-7000, -2000])
plt.show()
```

```python
plt.plot(x, L)
plt.ylim([-2910, -2900])
plt.xlim([-0.35, -0.25])
plt.show()

arg_max_2 = x[np.where(L == max(L))]
print("Значение максимума функции достигается в точке ", arg_max_2)
```
